```python
import csv
from matplotlib import pyplot as plt
```

```python
file_path = './Crash_Reporting_-_Drivers_Data.csv'
x_name = 'Latitude'
y_name = 'Longitude'

x = []
y = []
```

```python
with open(file_path) as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        x.append((float) (row[x_name]))
        y.append((float) (row[y_name]))
```

```python
plt.scatter(y, x, s = 10)
plt.xlabel('Longitude', fontsize=10)
plt.ylabel('Latitude', fontsize=10)
```

```python
times = []
want = 'Crash Date/Time'


with open(file_path) as file:
    reader = csv.DictReader(file)
    
    hours = ['12 am', '1 am', '2 am', '3 am', '4 am', '5 am', '6 am', '7 am', '8 am', '9 am', '10 am', '11 am', '12 pm', '1 pm', '2 pm', '3 pm', '4 pm', '5 pm', '6 pm', '7 pm', '8 pm', '9 pm', '10 pm', '11 pm']
    freqs = []
    for i in range(24):
        freqs.append(0)
        
    for row in reader:
            
        period = row[want][-2:]
        hour = (int) (row[want][-11:-9])
        if hour == 12:
            if (period == 'AM'):
                freqs[0] += 1
            else:
                freqs[12] += 1
        else:
            if (period == 'PM'):
                freqs[hour + 12] += 1
            else:
                freqs[hour] += 1
                
        
        
        
plt.bar(hours, freqs)
plt.xlabel('Time', fontsize=10)
plt.ylabel('Occurences', fontsize=10)
plt.rcParams["figure.figsize"] = [20, 10]
```
